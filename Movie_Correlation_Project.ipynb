```python
# Import Libraries

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.pyplot.rcParams['figure.figsize'] = (12,8)  # Adjust the size of the plots
```

```python
# Read the data

df = pd.read_csv(r'C:\Users\chatu\OneDrive\Documents\Projects\Python Projects\Data Files\movies.csv')
```

```python
# Look at the data

df.head()
```

```python
# Analyze for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for columns

df.dtypes
```

```python
# Change data type for columns filling missing values with zeros

df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int)
```

```python
# Look at the data

df.head()
```

```python
# Create correct year column

df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
# Look at the data

df.head()
```

```python
# Check for the highest grossing movie

df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Set Console to show all Data

pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df.head()['company'].drop_duplicates().sort_values(ascending = False)
```

```python
# Look at the data

df.head()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Movie')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Plot budget vs gross using Seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {'color': '#87CEEB'}, line_kws = {'color': '#00008B'})
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Movie')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Analyse Correlation
```

```python
df.corr(numeric_only = True, method = 'pearson')
```

```python
correlation_matrix = df.corr(numeric_only = True, method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Details')
plt.ylabel('Movie Details')
plt.show()
```

```python
# Analyse company

df.head()
```

```python
# Convert non-numeric columns to numeric

df_numeric = df.copy()

for col_name in df_numeric:
    if (df_numeric[col_name].dtype == 'object'):
        df_numeric[col_name] = df_numeric[col_name].astype('category')
        df_numeric[col_name] = df_numeric[col_name].cat.codes

df_numeric.head()
```

```python
correlation_matrix = df_numeric.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Details')
plt.ylabel('Movie Details')
plt.show()
```

```python
df_numeric.corr()
```

```python
corr_arr = df_numeric.corr()
corr_pairs = corr_arr.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Check for High Correlation Details

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
